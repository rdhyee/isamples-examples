```python
import httpx


# Get the vocabulary from the iSamples Central API
vocabulary_urls = {
    'material_sample': 'https://central.isample.xyz/isamples_central/vocabulary/material_sample_type',
    'features': 'https://central.isample.xyz/isamples_central/vocabulary/sampled_feature_type',
    'material_type': 'https://central.isample.xyz/isamples_central/vocabulary/material_type'
}

# Get the vocabulary from the iSamples Central API
vocabulary = {}
for key, url in vocabulary_urls.items():
    response = httpx.get(url)
    if response.status_code == 200:
        vocabulary[key] = response.json()
    else:
        print(f'Failed to get vocabulary for {key}')
        
```

```python
def walk(d, indent=0):
    for (k, v) in d.items():
        # print(k,type(v), v.keys() if isinstance(v, dict) else '')
        assert isinstance(v, dict) or isinstance(v, list)
        # look for label 
        assert 'label' in v
        print('  ' * indent, k, v['label']['en'])
        # look for children
        if 'children' in v:
            for child in v['children']:
                walk(child, indent+1)
```
